```python
import os
import glob
import json

# Directory containing the files
directory = r"/Users/marushikamanohar/Downloads/CIP/datasets/GridSybil_0709/VeReMi_25200_28800_2025-11-15_13:57:9/"

# Get all JSONL files matching the pattern
files = glob.glob(os.path.join(directory, "traceJSON-*.json"))

for file_path in files:
    filename = os.path.basename(file_path)
    
    # Extract vehicle number and module number from filename
    # Example filename: traceJSON-9-7-A0-25202-7.json
    parts = filename.split('-')
    vehicle_no = int(parts[1])
    module_no = int(parts[2])
    
    # Read JSONL line by line
    with open(file_path, 'r') as f:
        for line_number, line in enumerate(f, start=1):
            try:
                msg = json.loads(line)
            except json.JSONDecodeError:
                print(f"Error decoding JSON on line {line_number} in {filename}")
                continue
            
            sender = msg.get("sender")
            if sender == vehicle_no or sender == module_no:
                print(f"In file {filename}, line {line_number}: sender {sender} matches vehicle_no ({vehicle_no}) or module_no ({module_no})")
```

```python
import os
import glob
import json

root_dir = r"/Users/marushikamanohar/Downloads/CIP/datasets/GridSybil_0709/"

target_sender = 9
matches = []

# Walk through all subfolders
for dirpath, dirnames, filenames in os.walk(root_dir):
    for filename in filenames:
        if not filename.startswith("traceJSON-") or not filename.endswith(".json"):
            continue
        
        file_path = os.path.join(dirpath, filename)
        with open(file_path, "r") as f:
            for line_number, line in enumerate(f, start=1):
                try:
                    msg = json.loads(line)
                except json.JSONDecodeError:
                    continue
                
                sender = msg.get("sender")
                # Check both int and string cases
                if sender == target_sender or str(sender) == str(target_sender):
                    matches.append((file_path, line_number, sender))

# Print results
if matches:
    print(f"Found {len(matches)} occurrences of sender == {target_sender}:")
    for file_path, line, sender in matches:
        print(f"  {file_path}  (line {line})  sender={sender}")
else:
    print(f"No occurrences of sender == {target_sender} found in the entire dataset.")
```

```python
import os
import json
import glob

# Folder containing ground truth files
folder = r"/Users/marushikamanohar/Downloads/CIP/datasets/DataReplaySybil_0709"

# Step 1: Extract module numbers from ground truth filenames
module_numbers_from_filenames = set()
ground_truth_files = glob.glob(os.path.join(folder, "**", "traceGroundTruthJSON-*.json"), recursive=True)

for file_path in ground_truth_files:
    filename = os.path.basename(file_path)
    # Example: traceGroundTruthJSON-7.json -> module_no = 7
    parts = filename.split('-')
    if len(parts) >= 2:
        module_part = parts[1].split('.')[0]  # remove .json
        try:
            module_no = int(module_part)
            module_numbers_from_filenames.add(module_no)
        except ValueError:
            continue

print(f"Module numbers from filenames ({len(module_numbers_from_filenames)}): {sorted(module_numbers_from_filenames)}")

# Step 2: Extract unique senders from ground truth JSONL files
unique_senders = set()

for file_path in ground_truth_files:
    with open(file_path, 'r') as f:
        for line in f:
            try:
                msg = json.loads(line)
            except json.JSONDecodeError:
                continue
            sender = msg.get("sender")
            if sender is not None:
                unique_senders.add(int(sender))  # convert to int for comparison

print(f"Unique senders from ground truth files ({len(unique_senders)}): {sorted(unique_senders)}")

# Step 3: Compare the sets
if module_numbers_from_filenames == unique_senders:
    print(" Module numbers from filenames MATCH the unique senders in ground truth files!")
else:
    missing_in_senders = module_numbers_from_filenames - unique_senders
    missing_in_filenames = unique_senders - module_numbers_from_filenames
    print(" Sets do NOT match!")
    if missing_in_senders:
        print(f"Module numbers in filenames but not in senders: {sorted(missing_in_senders)}")
    if missing_in_filenames:
        print(f"Senders in ground truth but not in filenames: {sorted(missing_in_filenames)}")
```

```python
import os
import json
import glob

# Folders
logs_folder = r"/Users/marushikamanohar/Downloads/CIP/datasets/DataReplaySybil_0709"
ground_truth_folder = logs_folder  # assuming same parent folder

# Step 1: Extract module numbers from traceJSON filenames
module_numbers = set()
trace_files = glob.glob(os.path.join(logs_folder, "**", "traceJSON-*.json"), recursive=True)

for file_path in trace_files:
    filename = os.path.basename(file_path)
    parts = filename.split('-')
    if len(parts) >= 3:
        try:
            module_no = int(parts[2])  # second number is module_no
            module_numbers.add(module_no)
        except ValueError:
            continue

print(f"Module numbers from traceJSON files ({len(module_numbers)}): {sorted(module_numbers)}")

# Step 2: Extract unique senders from ground-truth files
unique_senders = set()
ground_truth_files = glob.glob(os.path.join(ground_truth_folder, "**", "traceGroundTruthJSON-*.json"), recursive=True)

for file_path in ground_truth_files:
    with open(file_path, 'r') as f:
        for line in f:
            try:
                msg = json.loads(line)
            except json.JSONDecodeError:
                continue
            sender = msg.get("sender")
            if sender is not None:
                try:
                    unique_senders.add(int(sender))
                except ValueError:
                    continue

print(f"Unique senders from ground-truth files ({len(unique_senders)}): {sorted(unique_senders)}")

# Step 3: Compare module numbers with unique senders
missing_in_senders = module_numbers - unique_senders
missing_in_modules = unique_senders - module_numbers

if not missing_in_senders and not missing_in_modules:
    print(" All module numbers match unique senders from ground-truth files!")
else:
    print(" Mismatches found:")
    if missing_in_senders:
        print(f"Module numbers in traceJSON files but NOT in ground-truth senders: {sorted(missing_in_senders)}")
    if missing_in_modules:
        print(f"Senders in ground-truth files but NOT in traceJSON module numbers: {sorted(missing_in_modules)}")
```
